## DECIDING FAETURE OF SPLITING

```python
predictors = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']
target = data['medv']
```

## SPLITING DATASET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.30)
```

```python
X_train
```

## SELF MADE FUNCTION FOR DEFINING ACCURACY

```python
def accuarcy(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train[predictors], y_train)
    y_pred = model.predict(X_test[predictors])
    ypred_train = model.predict(X_train[predictors])
    print ('MSE train', mean_absolute_error(y_train, ypred_train))
    print ('MSE test', mean_absolute_error(y_test, y_pred))
    r_2 = model.score(X_test[predictors], y_test)
    print ('R^2 test', r_2)
```

## MODEL BUILDING

```python
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression

svr_rbf = SVR(kernel='rbf', gamma=0.1)
svr_lineal = SVR(kernel='linear')
lr = LinearRegression()
```

## ACCUARCY / MSE / R^2

```python
print("----------")
print( "SVR - RBF")
print("----------")
accuarcy('SVR - RBF', svr_rbf, X_train, y_train, X_test, y_test)
print("----------")
print ('SVR - lineal')
print ('-----------')
accuarcy('SVR - lineal', svr_lineal, X_train, y_train, X_test, y_test)
print ('-----------')

print ('Regresion lineal')
print ('-----------')
accuarcy('Regresion lineal', lr, X_train, y_train, X_test, y_test)
```

## STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train[predictors])
X_train[predictors] = scaler.transform(X_train[predictors])
X_test[predictors] = scaler.transform(X_test[predictors])
```
